### Question 2

Given that the derivative security can be hedged, there exist a portfolio value process $X(t)$ with initial condition $X(0)$ satisfies
$$
\begin{align*}
X(T) = V(T)
\end{align*}
$$
Assume there are 2 risk-neutral probability measures $\mathbb {\tilde P}_{1}$ and $\mathbb {\tilde P}_{2}$, then the discounted portfolio value probability $D(t)X(t)$ is a martingale under both  $\mathbb {\tilde P}_{1}$ and $\mathbb {\tilde P}_{2}$. so we have
$$
\begin{align*}
V_{1}(t) &= \frac{1}{D(t)}\mathbb{\tilde E}_{1}[D(T)V(T)|\mathcal{F}(t)] = \frac{1}{D(t)}\mathbb{\tilde E}_{1}[D(T)X(T)|\mathcal{F}(t)] = X(t)\\
&=\frac{1}{D(t)}\mathbb{\tilde E}_{2}[D(T)X(T)|\mathcal{F}(t)] = \frac{1}{D(t)}\mathbb{\tilde E}_{2}[D(T)V(T)|\mathcal{F}(t)] = V_{2}(t)
\end{align*}
$$
So we can conclude that if $V(t)$ can be hedged, then its price under any risk-neutral probability measure is the same.

```python
import numpy as np
import matplotlib.pyplot as plt

def heston_model_paths(S0, V0, p, dt, T, num_paths):
    N = int(T/dt)
    paths = np.zeros((N+1, num_paths))
    paths[0] = S0
    
    for i in range(num_paths):
        S = S0
        V = V0
        for j in range(N):
            Be1 = np.random.normal(0, np.sqrt(dt))
            Be2 = np.random.normal(0, np.sqrt(dt))
            Wf = np.sqrt(3)*Be1 - Be2
            
            # Milstein scheme for the CIR process
            k = 1.2
            theta = 0.04
            xi = 0.3
            dV = k * (theta - V) * dt + xi * p * V**0.5 * Be2 + 0.25 * xi**2 * p**2 * (Be2**2 - dt)
            
            # Update the process V(t)
            V += dV
            
            # Calculate dln(S(t))
            dlnS = (0.05 - 0.5 * V) * dt + p * np.sqrt(V) * Wf - 0.15 * dt
            
            # Update the stock price
            S *= np.exp(dlnS)
            
            paths[j+1, i] = S
            
    return paths

# Parameters
S0 = 100
V0 = 0.04
p = 1  # You might want to calibrate the value of p
dt = 0.01  # Time step
T = 1  # Time horizon

# Simulate paths
paths = heston_model_paths(S0, V0, p, dt, T, 10)

# Plot the paths
for i in range(paths.shape[1]):
    plt.plot(np.linspace(0, T, int(T/dt)+1), paths[:, i])

plt.title('Heston Model Paths')
plt.xlabel('Time')
plt.ylabel('Stock Price')
plt.show()

```
