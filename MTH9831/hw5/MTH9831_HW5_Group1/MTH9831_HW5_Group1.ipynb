## Question 1

#### (a)

Since we have $dW(t) = \sum_{i=1}^n \cfrac{B_i(t)}{\sqrt{S(t)}}dB_i(t)$, using the fact that covariations of different components of Brownian motion are zero, we have
$$
[W]_t = \sum_{i=1}^n (\cfrac{B_i(t)}{\sqrt{S(t)}})^2 [B_i]_t = \sum_{i=1}^n \cfrac{B_i(t)^2}{S(t)} t = \cfrac{1}{S(t)} \sum_{i=1}^n B_i(t)^2 t = t
$$
Since $W(0) = 0$ and $W(t)$ is a continuous local martingale with quadratic variation equal to t, we have $W(t)$ is a standard brownian motion by Levy's characterization theorem.

#### (b)

Apply Ito's formula to $f(t,x)=x^2$, where x is the n-dimensional Brownian motion, we have
$$
dS(t) = 2\sum_{i=1}^n B_i(t)dB_i(t) + \sum_{i=1}^n dt = 2\sum_{i=1}^n B_i(t)dB_i(t) + n dt
$$
Substitute the expression for $dW(t)$ we have 
$$
dS(t) = 2\sum_{i=1}^n B_i(t)dB_i(t) + n dt = 2\sqrt{S(t)}dW(t) + n dt
$$

In CIR model $dr(t) = (\alpha-\beta r(t))dt+\sigma \sqrt{r(t)} dW(t)$, setting $\alpha=n,\beta=0, \sigma=\sqrt{2}$ we can recover the squared Bessel process.

#### (c)

Apply Ito's formula to $g(t,x)=\sqrt{x}$, where x is the sqaure Bessel process $S(t)$, we have
\begin{align*}
d\sqrt{S(t)} &= \cfrac{1}{2\sqrt{S(t)}}dS(t) - \cfrac{1}{8S(t)^{3/2}}d[S]_t = \cfrac{1}{2\sqrt{S(t)}}dS(t) - \cfrac{1}{2S(t)^{1/2}}dt \\
&= \cfrac{1}{2\sqrt{S(t)}}(2\sqrt{S(t)}dW(t) + ndt) - \cfrac{1}{2S(t)^{1/2}}dt \\
&= dW(t) + \cfrac{n-1}{2\sqrt{S(t)}}dt
\end{align*}
Hence we finish the proof.

## Question 7

(7) (Visualizing Girsanov’s formula) Let B'(t) = B(t) + t, t ∈ [0, 1].
(a) Generate a sample S of 100,000 paths of B'(t), t ∈ [0, 1] using 0.01 discretization.

(b) Using the function numpy.random.choice, sample 1,000 paths in S not uniformly but so that the weight of each path B'(t), t ∈ [0, 1], is proportional to Z(1) = e^(−B'(1)−1/2) . You need to normalize the weights so that their sum is 1. Let us call this new sample S'.

(c) Draw the histogram of B'(1/2) on the sample S'. It should look like a normal density with mean
0 and variance 1/2.

(d) Plot the first 10 paths from S. Plot the first 10 paths from S'.

```python
import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
n = 100000 # number of paths
m = 100 # number of time steps
dt = 0.01 # time step size
t = np.linspace(0, 1, m+1) # time grid

# Generate standard normal increments
dW = np.sqrt(dt) * np.random.randn(n, m)

# Initialize the paths matrix
B = np.zeros((n, m+1))

# Generate the paths of B(t) using Euler-Maruyama method
for i in range(1, m+1):
    B[:, i] = B[:, i-1] + dW[:, i-1]

# Define B'(t) = B(t) + t
B_prime = B + t
```

```python
# Define Z(1) as a function of B'(1)
def Z(B_prime_1):
    return np.exp(-B_prime_1 - 0.5)

# Compute the weights of each path
weights = Z(B_prime[:, -1])

# Normalize the weights so that their sum is 1
weights = weights / np.sum(weights)

# Use numpy.random.choice to sample 1000 paths with the given weights
indices = np.random.choice(n, size=1000, replace=False, p=weights)

# Define the new sample S'
S_prime = B_prime[indices, :]
```

```python
# Extract the values of B'(1/2) from S'
B_prime_half = S_prime[:, m//2]

# Plot the histogram with 50 bins and density=True
plt.figure(figsize=(8, 6))
plt.hist(B_prime_half, bins=50, density=True)
plt.xlabel("B'(1/2)")
plt.ylabel("Density")
plt.title("Histogram of B'(1/2) on S'")
plt.show()
```

```python
# Plot the first 10 paths from S
plt.figure(figsize=(8, 6))
plt.plot(t, B_prime[:10, :].T)
plt.xlabel('t')
plt.ylabel("B'(t)")
plt.title("First 10 paths from S")
plt.show()

# Plot the first 10 paths from S'
plt.figure(figsize=(8, 6))
plt.plot(t, S_prime[:10, :].T)
plt.xlabel('t')
plt.ylabel("B'(t)")
plt.title("First 10 paths from S'")
plt.show()
```
